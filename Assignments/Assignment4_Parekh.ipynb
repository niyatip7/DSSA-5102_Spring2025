Now that we have practiced, let's solve this murder!

All we know at the start of our investigation is the <b>murder</b> occurred on <b>Jan.15, 2018</b> and that it took place in <b>SQL City<b>.

```python
#Write a query in SQL to view the crime scene report 
#for the murder occurred on Jan.15, 2018 and that it took place in SQL City.
#Select ALL records FROM the table WHERE...


pd.set_option('display.max_colwidth', None)  # Show full text in each cell

#make a connection to the database using the engine
with engine.connect() as connection:
    #defining the query to use the crime scene report where the date of Jan 15 2018, crime was murder and took place in SQL city 
    #text () means it is a SQL theory 
    query = text("""SELECT * 
                FROM crime_scene_report 
                WHERE date='20180115'
                AND  type='murder'
                AND city='SQL City';
                    """)
    #perform the query and store them in a pandas dataframe 
    murder = pd.read_sql(query, connection)

#print
murder
#the murder that needs to be solved has 2 witnesses 
```

Another clue! It looks like we have two witnesses! Now who are they and how do we find them?

```python
#Write two queries in SQL to find each witness in the person table. 
#For the first query as first_witness, select all the records from person where the address_street_name is Northwestern Drive and order by adress_number
#Fetch the first_witness

#make a connection to the database using the engine
with engine.connect() as connection:
    #defining the query to use the person table and select where address_street_name is Northwestern Dr and order by address number
    #to find the last one 
    #text () means it is a SQL theory 
    query = text("""SELECT * 
                FROM person 
                WHERE address_street_name='Northwestern Dr'
                ORDER by address_number
                    """)
 #perform the query and store them in a pandas dataframe
    first_witness = pd.read_sql(query, connection)

#print 
first_witness
#is the first witness Morty Schapiro
#ID is 14887
```

```python
#For the third query as second_witness,
#Select all the records from person where the address_street_name is Frnaklin Ave and the name includes Annabel. 
#Fetch the second_witness

#make a connection to the database using the engine
with engine.connect() as connection:
    #defining the query to use the person table and select where address_street_name is Franklin Ave and order by address number 
    #text () means it is a SQL theory 
    query = text("""SELECT * 
                FROM person 
                WHERE address_street_name='Franklin Ave'
                AND name LIKE '%Annabel%';
                    """)
 #perform the query and store them in a pandas dataframe
    second_witness = pd.read_sql(query, connection)

#print 
second_witness
```
